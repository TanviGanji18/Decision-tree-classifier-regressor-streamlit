```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
x = pd.DataFrame(iris.data, columns=iris.feature_names)
x
```

```python
y = iris['target']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# applying decision tree regression
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(
    max_depth=4,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features=None,
    random_state=42
)
```

```python
model.fit(x_train, y_train)
```

```python
# visualization of the decision tree regression
from sklearn.tree import plot_tree
plt.figure(figsize=(12, 8))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names, max_depth=3)
plt.show()
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
params = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],
    'splitter': ['best', 'random'],
    'max_depth': [2, 3, 4, 5, 6, 1],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(
    estimator=DecisionTreeRegressor(random_state=42),
    param_grid=params,
    cv=5,
    scoring='r2'
)
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train, y_train)
```

```python
print("Best Params:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_

y_pred = best_model.predict(x_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```
