- entropy(measures impurity or randomness in a dataset,used in ID3)
- gini index(measures the probability of misclassification,used in CART)
##### what feature you need to select for splitting?
- Information gain(measures how much entropy decreases after a split.)

#### entropy(0,1)-->pure split-0,impure split-1
#### gini index(0.5 )-->pure split-0,impure split-0.5

# Decision Tree Classifier


```python
import numpy as np
from sklearn.datasets import load_iris
iris = load_iris()
print(iris.DESCR)
```

```python
print(iris['target'])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = pd.DataFrame(iris.data, columns=[
    'sepal length (cm)',
    'sepal width (cm)',
    'petal length (cm)',
    'petal width (cm)'
])
x
```

```python
y = iris['target']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
## applying decision tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train, y_train)
```

```python
# visualization of decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,
               filled=True,
               feature_names=x.columns,
               class_names=iris['target_names'],
               max_depth=3
               )
plt.show()
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
params = {
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [1, 2, 3, 4, 5, 6],
    'splitter': ['best', 'random'],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(treeclassifier, param_grid=params, cv=5,scoring='accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(x_test)
```

```python
y_pred
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
print(score)
```
